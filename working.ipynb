# Feature Engineering

```python
# Drop ID
df = df.drop(columns=["Student_ID"])

# Fill missing numerical values
num_cols = [
    "Family_Income",
    "Study_Hours_per_Day"
]

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill categorical
df["Parental_Education"] = df["Parental_Education"].fillna(
    df["Parental_Education"].mode()[0]
)
```

```python
y = df["Stress_Index"]

X = df.drop(columns=["Stress_Index"])
```

# Engineering different features
- Academic features
- Academic consistency
- Study Efficiency
- academic delay
-  academic load

```python
X["GPA_Trend"] = X["Semester_GPA"] - X["GPA"]
```

```python
X["Academic_Consistency"] = abs(
    X["Semester_GPA"] - X["CGPA"]
)
```

```python
X["Study_Efficiency"] = (
    X["Study_Hours_per_Day"] *
    X["Attendance_Rate"] / 100
)
```

```python
X["Delay_per_Hour"] = (
    X["Assignment_Delay_Days"] /
    (X["Study_Hours_per_Day"] + 1)
)
```

## Academic Load
### Standardize the contributing variables first, then combine them

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

load_cols = [
    "Assignment_Delay_Days",
    "Study_Hours_per_Day",
    "Attendance_Rate"
]

scaled = scaler.fit_transform(X[load_cols])

scaled = pd.DataFrame(
    scaled,
    columns=load_cols,
    index=X.index
)

X["Academic_Load"] = (
    scaled["Study_Hours_per_Day"]
    + scaled["Assignment_Delay_Days"]
    - scaled["Attendance_Rate"]
)
```

## More on Feature Engineering
- financial risk
- long commute
- working student
- scholarship support
- Access to internet
- semester
- overall engagements(weights0)
- GPA difference

```python
median_income = X["Family_Income"].median()

X["Financial_Risk"] = (
    X["Family_Income"] < median_income
).astype(int)
```

```python
median_time = X["Travel_Time_Minutes"].median()

X["Long_Commute"] = (
    X["Travel_Time_Minutes"] > median_time
).astype(int)
```

```python
X["Working_Student"] = (
    X["Part_Time_Job"] == "Yes"
).astype(int)
```

```python
X["Has_Scholarship"] = (
    X["Scholarship"] == "Yes"
).astype(int)
```

```python
#semester
X["Year of study"] = X["Year of study"].astype(str)
```

```python
X["Internet"] = (
    X["Internet_Access"] == "Yes"
).astype(int)
```

```python
X["Engagement"] = (
    0.7 * X["Attendance_Rate"] +
    0.3 * X["Study_Hours_per_Day"] * 10
)
```

```python
X["CGPA_Gap"] = (
    X["CGPA"] - X["Semester_GPA"]
)
```
